```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('graduate-admissions/Admission_Predict.csv')
```

```python
data
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(x = data['TOEFL Score'], y = data['Chance of Admit '])
plt.xlabel('TEOFL')
plt.ylabel('Admint')
plt.title('TEOFL vs. Admittance')
plt.suptitle('Figure 1 and 2')

plt.subplot(1, 2, 2)
plt.scatter(x = data['GRE Score'], y = data['Chance of Admit '], color='red')
plt.ylabel('Admint')
plt.xlabel('GRE')
plt.title('GRE vs. Admittance')

plt.show()
```

The graphics 1 and 2 shows the relationship between the respective test scores and the strength or possibility to enter a University.

```python
plt.figure(figsize=(9, 4))
plt.suptitle('Figure 3 and 4')

plt.subplot(1, 2, 1)
plt.hist(data['TOEFL Score'], color='blue', bins=40)
plt.xlabel('TOELF')
plt.title('TOEFL Histogram')

plt.subplot(1, 2, 2)
plt.hist(data['GRE Score'], bins=40)
plt.xlabel('GRE')
plt.title('GRE Histogram')

plt.tight_layout()
plt.show()
```

The graphics 3 and 4 shows the TOEFL, and GRE histogram distribution. The graphics show TOEFL to concentrate more incident amount of students at the 110 scores and the GRE the distribution is more distributed.

```python
plt.boxplot(data['University Rating'], vert=False)
plt.title('University Rating Boxplot')
plt.xlabel('Values')
df = pd.DataFrame(data, index=['University Rating'])
df.T.boxplot(vert=False)
plt.show()
```

The University Rating Boxplot demonstrate the range of distribution of the university rating value is symmetrical, the Interquartile Range (IQR) is 2, and the media is 3.

```python
plt.figure(figsize=(9, 4))

plt.subplot(1,2,1)
plt.scatter(x = data['TOEFL Score'], y = data['University Rating'])
plt.ylim(0.5, 5.5)
plt.ylabel('Rating')
plt.xlabel('Score')
plt.title("TOEFL Score vs. University Rating")

plt.subplot(1,2,2)
plt.scatter(x = data['GRE Score'], y = data['University Rating'])
plt.ylim(0.5, 5.5)
plt.ylabel('Rating')
plt.xlabel('Score')
plt.title("GRE Score vs. University Rating")

plt.show()
```

The correlation of University Rating has not a clear relationship with the level of the score of TOEFL or GRE. The extreme scores (low and high) are the only ones that could predict what level of University the student could get access to.
